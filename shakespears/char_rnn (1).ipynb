```python
import requests
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
response=requests.get(url)
with open('tiny_shakespeare.txt', 'w', encoding='utf-8') as f:
    f.write(response.text)
with open('tiny_shakespeare.txt') as f:
    shakespeare_text = f.read()
```

```python
print(shakespeare_text[:148])
```

```python
tokenizer=keras.preprocessing.text.Tokenizer(char_level=True)
tokenizer.fit_on_texts(shakespeare_text)
```

```python
tokenizer.texts_to_sequences(["First"])
```

```python
tokenizer.sequences_to_texts([[20, 6, 9, 8, 3]])
```

```python
max_id=len(tokenizer.word_index)
dataset_size=tokenizer.document_count
```

```python
[encoded]=np.array(tokenizer.texts_to_sequences([shakespeare_text]))-1
train_size=dataset_size*90//100
dataset=tf.data.Dataset.from_tensor_slices(encoded[:train_size])
```

```python
n_steps=100
window_length=n_steps+1
dataset=dataset.window(window_length,shift=1,drop_remainder=True)
```

```python
dataset=dataset.flat_map(lambda window:window.batch(window_length))
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
batch_size=32
dataset=dataset.shuffle(10000).batch(batch_size)
dataset=dataset.map(lambda windows:(windows[:,:-1],windows[:,1:]))
```

```python
dataset=dataset.map(
    lambda X_batch,Y_batch:(tf.one_hot(X_batch,depth=max_id),Y_batch)
)
```

```python
dataset=dataset.prefetch(1)
```

```python
for X_batch,Y_batch in dataset.take(1):
    print(X_batch.shape,Y_batch.shape)
```

```python
model = keras.models.Sequential([
    keras.layers.GRU(128, return_sequences=True, input_shape=[None, max_id],
                     #dropout=0.2, recurrent_dropout=0.2),
                     dropout=0.2),
    keras.layers.GRU(128, return_sequences=True,
                     #dropout=0.2, recurrent_dropout=0.2),
                     dropout=0.2),
    keras.layers.TimeDistributed(keras.layers.Dense(max_id,
                                                    activation="softmax"))
])
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam")
history = model.fit(dataset, epochs=10)
```

I just run 5 epochs you can run more if you want it will lead to better result but you can get much better results in stateful rnn .

```python
def preprocess(texts):
    X = np.array(tokenizer.texts_to_sequences(texts)) - 1
    return tf.one_hot(X, max_id)
```

```python
X_new = preprocess(["How are yo"])
Y_pred = np.argmax(model(X_new), axis=-1)
tokenizer.sequences_to_texts(Y_pred + 1)[0][-1]
```

```python
tf.random.set_seed(42)

tf.random.categorical([[np.log(0.5), np.log(0.4), np.log(0.1)]], num_samples=40).numpy()
```

```python
def next_char(text, temperature=1):
    X_new = preprocess([text])
    y_proba = model(X_new)[0, -1:, :]
    rescaled_logits = tf.math.log(y_proba) / temperature
    char_id = tf.random.categorical(rescaled_logits, num_samples=1) + 1
    return tokenizer.sequences_to_texts(char_id.numpy())[0]
```

```python
tf.random.set_seed(42)

next_char("How are yo", temperature=1)
```

```python
def complete_text(text, n_chars=500, temperature=1):
    for _ in range(n_chars):
        text += next_char(text, temperature)
    return text
```
